```python
import pandas as pd
from utils.plot_utils import plot_mean_std_fill
import matplotlib.pyplot as plt
import numpy as np
```

```python
ga_perf=pd.read_csv("./perf_sphereGAOptimizer.csv")
pso_perf=pd.read_csv("./perf_spherePSOptimizer.csv")
nsgaII_perf=pd.read_csv("./perf_sphereNSGAIIOptimizer.csv")

ax=plot_mean_std_fill(ga_perf.fit_mean,ga_perf.fit_std,'g')


plot_mean_std_fill(pso_perf.fit_mean,pso_perf.fit_std,'r',ax)


plot_mean_std_fill(nsgaII_perf.fit_mean,nsgaII_perf.fit_std,'b',ax)

plt.plot([100,101],'k',ls="--")
plt.plot([100101],'k',ls="-")
plt.plot(ga_perf.fit_best,'g',ls='--')
plt.plot(pso_perf.fit_best,'r',ls='--')
plt.plot(nsgaII_perf.fit_best,'b',ls='--')
plt.legend(["GA","PSO","NSGAII","Mean","Best"])
plt.ylim([-125,10])
plt.xlim([-1,20])
plt.xticks(np.linspace(0, 20, num=5,endpoint=True))
plt.xlabel("Step")
plt.ylabel("Fitness")
plt.title("Sphere function optimisation")
plt.tight_layout()

plt.savefig("sphere_opti.pdf", dpi=None, facecolor='w', edgecolor='k',
    orientation='portrait', papertype=None, format="pdf",
    transparent=False, bbox_inches=None, pad_inches=0.1,
    frameon=True, metadata=None)
```
