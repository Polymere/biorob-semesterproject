```python
import pandas as pd
RAW_DATA_PATH="../../data/raw_example.csv"
RAW_REFERENCE_PATH="../../data/raw_reference.csv"
TIME_STEP=1.0 #ms
import matplotlib.pyplot as plt

data_in=pd.read_csv(open(RAW_DATA_PATH))
data_ref=pd.read_csv(open(RAW_REFERENCE_PATH))
def preprocess(data_in):
    data_out=pd.DataFrame(data_in.index)
    data_out["time"]=data_in.index*TIME_STEP
    data_out.set_index(data_out["time"],inplace=True)
    activation=data_in.filter(like="act",axis=1)
    data_out["energy"]=activation.sum(axis=1,skipna=True)
    return data_out
trial=preprocess(data_in)
ref=preprocess(data_ref)
```

```python
plt.plot(trial["energy"][0:200])
plt.plot(ref["energy"][0:200])
```
