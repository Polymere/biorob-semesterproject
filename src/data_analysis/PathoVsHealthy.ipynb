```python
import pandas as pd
import matplotlib.pyplot as plt
from utils.plot_utils import plot_mean_std_fill
from data_analysis.event_detection import interp_gaitprcent,interp_gaitprcent_df,get_rep_var_from_contact
import numpy as np
PATIENT1_PATH="../../data/patient1.csv"
PATIENT2_PATH="../../data/patient2.csv"
WINTERNORMAL_PATH="../../data/winter_data/data_normal.csv"
DIR_REF_CPP="../../data/ref_cpp/"
contact=pd.read_csv(DIR_REF_CPP+"footfall1",sep=" ")
joints=pd.read_csv(DIR_REF_CPP+"joints_angle1",sep=" ")
```

```python
df1=pd.read_csv(PATIENT1_PATH)
df2=pd.read_csv(PATIENT2_PATH)
df_win=pd.read_csv(WINTERNORMAL_PATH)
```

```python
interp1=interp_gaitprcent(-df1.LANKLE,100)
interp2=interp_gaitprcent(-df2.LANKLE,100)
interp_win=interp_gaitprcent(df_win.ankle,100)
plt.plot(interp1)
plt.plot(interp2)
plt.plot(interp_win)

metric='ANGLE_ANKLE_LEFT'
rep,var=get_rep_var_from_contact(contact,metric,joints)
plot_mean_std_fill(-rep*180/np.pi,var*180/np.pi,'r')

plt.legend(["Patient1","Patient2","Winter"])
plt.title("Left ankle");
plt.xlabel("stride %")
plt.ylabel("Deg");
```

```python
interp1r=interp_gaitprcent(-df1.RANKLE,100)# inversed again !
interp2r=interp_gaitprcent(-df2.RANKLE,100)
interp1l=interp_gaitprcent(-df1.LANKLE,100)
interp2l=interp_gaitprcent(-df2.LANKLE,100)

plt.plot(interp1r,'g',ls="-")
plt.plot(interp2r,'r',ls="-")

interp_win=interp_gaitprcent(-df_win.ankle,100)

plt.plot(interp_win)
plt.plot(np.linspace(-100,-101),'k',ls="--")
plt.plot(np.linspace(-100,-101),'k',ls="-")
plt.ylim([-20,50])
plt.xlim([0,100])

plt.plot(interp1l,'g',ls="--")
plt.plot(interp2l,'r',ls="--")

plt.legend(["Patient1","Patient2","Winter","Left  side","Right side"],ncol=2)
plt.title("Ankle");
plt.xlabel("stride %")
plt.ylabel("Angle [deg]");
plt.savefig("ankle_cmp.pdf", dpi=None, facecolor='w', edgecolor='k',
    orientation='portrait', papertype=None, format="pdf",
    transparent=False, bbox_inches=None, pad_inches=0.1,
    frameon=True, metadata=None)
```
